```python
import pandas as pd
import os

os.chdir("C:/Users/N/Projects/dataset")

legal_admin_mix = pd.read_excel("data/external/KIKmix.20240208.xlsx", dtype={'법정동코드':object, '행정동코드':object})

legal_admin_mix['법정동시군구코드'] = legal_admin_mix['법정동코드'].str[:5]
legal_admin_mix['법정동읍면동코드'] = legal_admin_mix['법정동코드'].str[5:]

legal_admin_mix = legal_admin_mix[legal_admin_mix['법정동읍면동코드'] != '00000']
legal_admin_mix = legal_admin_mix.reset_index(drop=True)

legal_admin_mix = legal_admin_mix[['행정동코드', '법정동코드']].drop_duplicates()

legal_admin_mix.head()
```

```python
legal_info_b = pd.read_csv("data/processed/legal_info_b.csv", dtype={'법정동코드':object})
admin_info_h = pd.read_csv("data/processed/admin_info_h.csv", dtype={'행정동코드':object})

legal_admin_mix_total = pd.merge(legal_admin_mix, legal_info_b, on='법정동코드', how='left')
legal_admin_mix_total = pd.merge(legal_admin_mix_total, admin_info_h, on='행정동코드', how='left')

legal_admin_mix_total.head(3).T
```

```python
# legal_admin_mix_total.to_csv('data/processed/legal_admin_mix.csv', index=False)
```

```python
legal_admin_mix_seoul = legal_admin_mix_total[legal_admin_mix_total['행정동코드'].astype('str').str.startswith('11')]

legal_admin_mix_seoul.to_csv('data/processed/legal_admin_mix_seoul.csv', index=False)
```
