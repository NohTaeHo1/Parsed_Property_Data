```python
import pandas as pd
import os
os.chdir("C:/Users/N/Projects/dataset")

admin_info_h = pd.read_excel("data/external/KIKcd_H.20240208.xlsx", dtype = {'행정동코드':object})

admin_info_h['행정동시군구코드'] = admin_info_h['행정동코드'].str[:5]
admin_info_h['행정동읍면동코드'] = admin_info_h['행정동코드'].str[5:]

admin_info_h = admin_info_h[admin_info_h['행정동읍면동코드'] != "00000"]
admin_info_h = admin_info_h[['행정동코드', '시도명', '시군구명', '읍면동명', '행정동시군구코드', '행정동읍면동코드']]

admin_info_h = admin_info_h.reset_index(drop = True)

admin_info_h = admin_info_h.where(pd.notnull(admin_info_h), " ")
admin_info_h['주소'] = admin_info_h['시도명']+" "+admin_info_h['시군구명']+" "+admin_info_h['읍면동명']

admin_info_h['주소'] = admin_info_h['주소'].str.replace('  ', ' ').str.strip()

admin_info_h.columns = ['행정동코드', '행정동시도명', '행정동시군구명', '행정동읍면동명', '행정동시군구코드', '행정동읍면동코드', '행정동주소']

admin_info_h.head()

# admin_info_h.to_csv("data/processed/admin_info_h.csv", index = False)
```

```python
admin_info_h_seoul = admin_info_h[admin_info_h['행정동코드'].astype(str).str.startswith('11')]

admin_info_h_seoul.to_csv("data/processed/admin_info_h_seoul.csv", index = False)
```
