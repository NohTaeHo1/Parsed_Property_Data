```python
import pandas as pd
import os
os.chdir("C:/Users/N/Projects/dataset")


legal_info_b = pd.read_excel("data/external/KIKcd_B.20240208.xlsx", dtype={'법정동코드':object})

legal_info_b.head()
```

```python
legal_info_b['법정동시군구코드'] = legal_info_b['법정동코드'].str[:5]
legal_info_b['법정동읍면동코드'] = legal_info_b['법정동코드'].str[5:]
legal_info_b.head()
```

```python
# 읍면동이 없는 경우를 제외
legal_info_b = legal_info_b[legal_info_b['법정동읍면동코드'] != "00000"]
legal_info_b = legal_info_b.reset_index(drop = True)
# 필요없는 컬럼 없애기
legal_info_b = legal_info_b[['법정동코드', '시도명', '시군구명', '읍면동명', '동리명', '법정동시군구코드', '법정동읍면동코드']]
legal_info_b.head()
```

```python
# where(cond, other), 여기서 cond는 조건이고, other는 조건을 만족하지 않는 요소를 대체할 값
legal_info_b = legal_info_b.where(pd.notnull(legal_info_b), "")

legal_info_b['주소'] = legal_info_b['시도명']+" "+legal_info_b['시군구명']+" "+legal_info_b['읍면동명']+" "+legal_info_b['동리명']

legal_info_b['주소'] = legal_info_b['주소'].str.replace('  ', ' ').str.strip()

legal_info_b.head()
```

```python
legal_info_b.to_csv("data/processed/legal_info_b.csv", index = False)
```
